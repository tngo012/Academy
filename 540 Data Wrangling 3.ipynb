```python
# Load required libraries
import pandas as pd
import numpy as np
```

```python
# Load the data into a dataframe
# use "Candy Hierarchy 2015 Survey Response" file for the data
df = pd.read_excel('CANDY-HIERARCHY-2015-SURVEY-Responses.xlsx')
df.head(7)
```

### At a quick glimpse at the first few rows, the 4th row has a missing data on the Butterfinger column.

### 1. Filter out missing data

```python
# check the columns
df.columns
```

```python
# drop all columns that have only missing data, as long as the columns have some values, they will be kept
df_1 = df.dropna(axis=1, how='all')
df_1
```

```python
# drop only rows with missing data, no columns or rows will have a missing value
df_1 = df_1.dropna()
df_1
```

### 2. Fill in missing values

```python
# before fill in the missing age
df.iloc[36,1]
```

```python
# the age column has missing values, replace the missing age with the previous known number
age = df.iloc[:,1]
age = age.fillna(method='ffill')
age
```

```python
# after fill in the missing age
age[36]
```
