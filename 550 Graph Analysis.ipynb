```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib
import matplotlib.pyplot as plt
```

```python
# load the data into a dataframe
data = pd.read_csv('mtcars.csv')
```

```python
data.head()
```

### 3A. Display the same analysis using 3 different charts (ex. A bar chart, a line chart and a pie chart)
### 3B. Use appropriate, complete, professional labeling.

```python
# assign variables from the data columns
cyl = data['cyl']
mpg = data['mpg']
model = data['model']
```

```python
# plot a horizontal bar chart of car mpg vs car cylinder
plt.rcdefaults()
fig, ax = plt.subplots()


ax.barh(cyl, mpg, align='center')
ax.set_yticks(cyl)
ax.set_yticklabels(cyl)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Miles per gallon')
ax.set_title('Fuel Efficiency of Different Cylinders')

plt.show()
```

```python
# Plot a line chart of car mpg versus car cylinder
fig = plt.figure()
ax = plt.axes()

plt.plot(mpg, cyl)
plt.title('Fuel Efficiency of Different Cylinders')
plt.xlabel('Miles per gallon')
plt.ylabel('Cylinder')
```

```python
# Plot a pie chart of car mpg and car models
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

plt.pie(mpg, labels=cyl,autopct='%1.f%%')
plt.legend(model)
plt.axis('equal')
plt.show()
```
