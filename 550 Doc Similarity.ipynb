```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
```

```python
# read the data from the DailyComments.csv file
# the link to the file is 
# https://www.kaggle.com/smid80/coronavirus-covid19-tweets#2020-03-28%20Coronavirus%20Tweets.CSV

dataframe = pd.read_csv('coronavirus_tweets.csv')
df_txt = dataframe['text']
print(df_txt[1:5000])
```

```python
# You must use some kind of vectorization method/tool 
# (my example uses sklearn count.vectorizer but you can use any vectorization tool or Jaccard Distance.

# the dataset has more than half a million lines, I choose only 5000 lines.
# I use the vectorization method to tokenize the words
corpus = df_txt[1:5000]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print("")
print("vectorized words")
print("")
print(vectorizer.get_feature_names())
print("")
print("Identify Feature Words - Matrix View")
print("")
print( X.toarray())
```

```python
df = pd.DataFrame({'text' : corpus})

# check for positive words and negative words
# attempt to add some words, which I think represent the positve and the negative tweets
df['positive1'] = df.text.str.count('good')
df['positive2']= df.text.str.count('great')
df['positive3']= df.text.str.count('safe')
df['positive4']= df.text.str.count('fine')
df['negative'] = df.text.str.count('bad')
df['negative2'] = df.text.str.count('approved')
df['negative3'] = df.text.str.count('negative')
df['negative4'] = df.text.str.count('dangerous')
df['TotScore'] = df.positive1 + df.positive2 + df.positive3 + df.positive4 - df.negative - df.negative2 - df.negative3 - df.negative4

print("")
print(df)

Z = sum(df['TotScore'])
print("")
print("Overall Score:  ",Z)
```

```python
# Create some kind of a dictionary of sample words you will use to search /categorize your data.
# jacard  - comparing words

import collections

def jaccard(a, b):
    c = a.intersection(b)
    print("this is intersection", c)
    d=a.union(b)
    print("this is union", d)
    _a = collections.Counter(a)
    _b = collections.Counter(b)
    c = (_a - _b) + (_b - _a)
    print("c", c)
    n = sum(c.values())
    print(_a)
    print(_b)
    print("_a and _b")
    print ("n", n)
    return float(len(c)) / (len(a) + len(b) - len(c))

list1 = ['government', 'patients', 'sick', 'drug', 'government', 'time']
list2 = ['government', 'patients', 'treat', 'effect', 'government', 'late']

words1 = set(list1)
words2 = set(list2)
print("words1:  ", words1)
print("words2:  ", words2)
z = jaccard(words1, words2)
print("this is z", z*100)

```

```python
# Chapter 7
# n-gram Feature Extraction

def ngrams(words, n=2):
    for idx in range(len(words)-n+1):
        yield tuple(words[idx:idx+n])
```
